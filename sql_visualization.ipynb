```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
correlation_survival_amount = pd.read_csv(r"C:\Users\Atefah\Desktop\IRONHACK\SQL_PROJECT\failure_factors_tables\corrrelation_survival_amount.csv")
```

```python
# Define custom ordering for funding levels
funding_order = ["<$1M", "$1M-$10M", "$10M-$50M", "$50M-$100M", "$100M-$500M", "$500M-$1B", ">$1B"]

# Ensure funding_level is a categorical variable with correct order
correlation_survival_amount["funding_level"] = pd.Categorical(
    correlation_survival_amount["funding_level"], 
    categories=funding_order, 
    ordered=True
)

# Sort the dataframe by the ordered funding levels
correlation_survival_amount = correlation_survival_amount.sort_values("funding_level")

# Plot line chart
plt.figure(figsize=(10, 5))
sns.lineplot(x="funding_level", y="avg_survival_years", data=correlation_survival_amount, marker="o", color="blue")

# Labels and title
plt.xlabel("Funding Level")
plt.ylabel("Average Survival Years")
plt.title("Average Survival Years by Funding Level")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Define the list to store dataframes
df_list = []

# Loop through each file and sector
for sector, file_path in files.items():
    # Read the CSV file
    df = pd.read_csv(file_path)
    
    # Check available columns
    print(f"Columns in {sector} dataset: {df.columns}")
    
    # Ensure 'end_year' exists, otherwise rename it
    if "end_year" not in df.columns:
        possible_year_columns = [col for col in df.columns if "year" in col.lower()]
        if possible_year_columns:
            df.rename(columns={possible_year_columns[0]: "end_year"}, inplace=True)
        else:
            print(f"Skipping {sector}: No year column found.")
            continue  # Skip this dataset if no year column is available
    
    # Select only 'end_year' and add a new column for 'Sector'
    df = df[["end_year"]].copy()
    df["Sector"] = sector  # Add sector name
    
    # Append to the list
    df_list.append(df)

# Combine all datasets
df_combined = pd.concat(df_list, ignore_index=True)

# Display sample data
print(df_combined.head())
```

```python
# Count occurrences of failures per year and sector
df_time_series = df_combined.groupby(["end_year", "Sector"]).size().reset_index(name="count")

# Display result
print(df_time_series.head())
```

```python
# Filter the data for end_year > 2000
df_filtered = df_time_series[df_time_series["end_year"] > 2000]

# Set figure size
plt.figure(figsize=(12, 6))

# Plot each sector separately
for sector in df_filtered["Sector"].unique():
    df_subset = df_filtered[df_filtered["Sector"] == sector]
    plt.plot(df_subset["end_year"], df_subset["count"], marker="o", label=sector)

# Formatting
plt.xlabel("Year")
plt.ylabel("Number of Failures")
plt.title("Time Series of Startup Failures by Sector (After 2000)")
plt.legend()
plt.grid(True)
plt.show()
```
